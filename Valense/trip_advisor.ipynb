```python
# Import Dependencies and Setup
import pandas as pd
import csv
import os

# Files to Load
museum1_path = os.path.join(os.getcwd(), "Raw_Data", "MuseumFile2018_File1_Nulls.csv")
museum2_path = os.path.join(os.getcwd(), "Raw_Data", "MuseumFile2018_File2_Nulls.csv")
museum3_path = os.path.join(os.getcwd(), "Raw_Data", "MuseumFile2018_File3_Nulls.csv")
# museums_path = os.path.join(os.getcwd(), "Raw_Data", "museums_separated.csv")
# trip_path = os.path.join(os.getcwd(), "Raw_Data", "tripadvisor_merged.csv")
trip1_path = os.path.join(os.getcwd(), "Raw_Data", "tripadvisor_museum_USonly.csv")
# trip2_path = os.path.join(os.getcwd(), "Raw_Data", "tripadvisor_museum_world.csv")
```

```python
# Read Files and Store Into Pandas Data Frame
museum1_df = pd.read_csv(museum1_path, encoding= 'unicode_escape')
museum2_df = pd.read_csv(museum2_path, encoding= 'unicode_escape')
museum3_df = pd.read_csv(museum3_path, encoding= 'unicode_escape')
# trip_df = pd.read_csv(trip_path)
trip1_df = pd.read_csv(trip1_path)
# trip2_df = pd.read_csv(trip2_path)
```
